```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('titanic2.csv', usecols=['Age','Pclass','SibSp','Parch','Survived'])
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
X=df.iloc[:,1:5]
y=df.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy', cv=7))
```

```python
X['Family']=X['SibSp'] + X['Parch'] +1 #concatenate the sibling spouse and parent child column and add 1, 
```

```python
X.head()
```

```python
# logic is to assign 0 who is travelling alone, 1 who is with small family, 2 who is travelling with large family
def myfunc(num):
    if num==1:
        return 0
    elif num>1 and num <=4:
        return 1
    else:
        return 2
```

```python
X['family_type']=X['Family'].apply(myfunc)
```

```python
X.sample(8)
```

```python
X.drop(columns=['SibSp','Parch','Family','is_with_fam'])
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy', cv=25))
```

```python
#accuracy improved by 1%
```

### Feature Splitting

```python
df=pd.read_csv('titanic2.csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
df['Name'].str.split(',',expand=True)
```

```python
df['Title']=df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```

```python
df.head()
```

```python
(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)
```

```python
print(df['Title'].unique())
df['Title'] = df['Title'].str.strip().str.replace('.', '').str.title()
```

```python
df['married']=0
df.loc[df['Title']=='Mrs','married']=1
```

```python
df.head()
```

```python
(df.groupby('married')['Survived'].mean()).sort_values(ascending=False)
```

```python
#married women survived more than unmarried women
```
