# scatterplot (num to num)

```python
#make a scatterplot with total bill on x axis and tip payed on y axis to understand the relationship between bills and tips
#method 1
# to distinguish sex of customers who payed tips , call hue in seaborn's scatterplot function
# to distinguish between smokers and non smokers using style, it shows crosses and circles refering to non-smokers and smokers respectively.

sns.scatterplot(data = tips, x ='total_bill', y ='tip', hue = tips['sex'], style = tips['smoker'], size = tips['size'])
```

```python
#method 2
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'])
```

# Distplot for num and cat

```python
import seaborn as sns
import pandas as pd
tit = pd.read_csv('titanic2.csv')
sns.distplot(tit[tit['Survived']==0]['Age'], hist = False)
sns.distplot(tit[tit['Survived']==1]['Age'], hist = False)
# comparing which passengers survived according to their ages
#pdf = probability density function
```
