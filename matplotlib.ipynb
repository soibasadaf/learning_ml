```python
df["exp"] = [2.2, 3.5, 2.5 , 4.5, 2.5, 3.5, 4, 3, 5]
```

```python
import matplotlib.pyplot as plt
```

# bubbleplot for multivariate analysis

```python
#plt.scatter(df['age'], df['salary'], c = df['dept'].map{'hr' : 'yellow', 'fin':'blue','it': 'green'}, labels ='hr', 'it', 'fin')

plt.title('age vs experience vs dept')

color = {'hr' : 'yellow', 'fin':'blue','it': 'green'}
            
for dept, color in color.items():
    df_dept = df[df['dept'] == dept]
    plt.scatter(df_dept['age'], df_dept['salary'], c = color, label = dept)


plt.xlabel('age')
plt.ylabel('salary')

plt.legend()

plt.show()
```

```python
plt.scatter(df['age'], df['salary'], s = df['exp']*12 , color = 'orange', edgecolors = 'black')
plt.title('age vs experience vs salary')
plt.xlabel('age')
plt.ylabel('salary')
plt.show()

```

# OO api

```python
fig, axis = plt.subplots(1,3, figsize =(15,7))


#lineplot

axis[0].plot(ages['age'], df['salary'], color = "blue" , marker = 'x', markersize = 2, linewidth = '2')
axis[0].grid()
axis[0].set_title('linear')
axis[0].set_xlabel('age')
axis[0].set_ylabel('salary')

#hist

axis[1].hist(df['salary'], bins = 5, color = 'green')
axis[1].set_title('hist')
axis[1].set_xlabel('age')
axis[1].set_ylabel('frequency')

#boxplot

axis[2].boxplot(df['salary'])
                 
axis[2].set_title('boxplot')
axis[2].set_xlabel('salary')

plt.savefig('sales_figure.png')
plt.show()
```
