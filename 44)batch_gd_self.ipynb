```python
class GDRegressor:

    def __init__(self,learning_rate=0.01,epochs=100):
        self.coef_ = None
        self.intercept_ = None
        self.lr= learning_rate
        self.epochs = epochs


    def fit(self,X_train,y_train):
        #initialize the coefficients
        self.intercept_= 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):

            y_hat=np.dot(X_train,self.coef_) + self.intercept_  #step 1 , i need ycap as in betas term
            der_intercept= -2*np.mean(y_train-y_hat) # step 2, subtract ycap with y orig , divide by n (mean), multiply by -2
            self.intercept_= self.intercept_ - self.lr*der_intercept # update slope , slope - lr*der-slope

            
            der_coef= -2*np.dot((y_train-y_hat),X_train)/X_train.shape[0] # y_train = y, y_hat=y-cap, x_train.shape[0]=#ofcolumns{n}
            self.coef_ = self.coef_- self.lr*der_coef 
            print(self.intercept_,self.coef_)
            


    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
gdr =GDRegressor()
```

```python
gdr = GDRegressor(epochs=1000,learning_rate=0.5)
```

```python
gdr.fit(X_train,y_train)
```

```python
# initialized intercept(scalar) with 0 and coef(betas_values) with 1 according to norms.
```

```python
y_pred = gdr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
