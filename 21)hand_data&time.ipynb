```python
#dates
date['date']=pd.to_datetime(date['date']) #to.datetime for dtype conversion
date.info()
```

```python
#to extract year
date['date_year']=date['date'].dt.year
date.sample(3)
```

```python
#extracting months
date['date_month']=date['date'].dt.month
date.sample(5)
```

```python
#month name
date['month_name']=date['date'].dt.month_name()
date.sample(5)
```

```python
#extracting days
date['date_day']=date['date'].dt.day
date.sample(5)
```

```python
#week days
date['week_day']=date['date'].dt.dayofweek
date.sample(5)
```

```python
# week days name
date['week_day_name']=date['date'].dt.day_name()
date.sample(3)
```

```python
#is thw day weekend
date['is_weekend']=np.where(date['week_day_name'].isin(['Saturday','Sunday']),1,0)
date.sample(5)
```
