# encoding categorical data

```python
# ordinal encoding[for features], label encoding[for labels] (converting categorical data into numeric data) 
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(8)
```

```python
# age is numerical
# gender is nominal categrical data{it need one hot encoding} remove it for now
df = df.iloc[:, 2:]
```

```python
df.sample(8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:2],
                                                   df.iloc[:, 2],
                                                   test_size=0.3)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train
```

```python
X_test
```

```python
# specify the order to encoder passing lists
order = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
order.fit(X_train)

```

```python
X_train_encoded=order.transform(X_train)
X_test_encoded=order.transform(X_test)
```

```python
X_train_encoded
```

```python
X_test_encoded
```

```python
order.categories_
```
